```python
import os
import glob
import pandas as pd
import numpy as np
import pyreadr

# your model folders and display names   NO_RF_five_fields
folders_titles = [
    ('half_RANN_outcome_one_field',             'Half R ANN 1 field'),
    ('half_RANN_outcome_five_fields',            'Half R ANN 5 fields'),
    ('half_RANN_outcome_ten_fields',            'Half R ANN 10 fields'),
    ('RANN_outcome_one_field',       'R ANN 1 field'),
    ('RANN_outcome_five_fields',      'R ANN 5 fields'),
    ('RANN_outcome_ten_fields',      'R ANN 10 fields'),
    #('Simple_ANN_1field',             'Simple ANN 1 field'),
    ('Simple_ANN_outcome_five_fields',            'Simple ANN 5 fields'),
    #('Simple_ANN_10fields',           'Simple ANN 10 fields'),
    ('HalfRRF_one_field',              'Half R RF 1 field'),
    ('HalfRRF_five_fields',             'Half R RF 5 fields'),
    ('HalfRRF_ten_fields',            'Half R RF 10 fields'),
    ('NO_RF_five_fields',            'NO_RF 5 fields'),
    ('NO_RF_ten_fields',            'NO_RF 10 fields'),
    ('CF_outcome_1_fields',                  'CF 1 field'),
    ('CF_outcome_5_fields',                'CF 5 field'),
    ('CF_outcome_10_fields',                 'CF 10 field'),
]

base_path = '/Users/mmousavi2/Dropbox/Causal_climate/Data_and_results_CF_Continuous/results'

records = []

for folder, title in folders_titles:
    folder_path = os.path.join(base_path, folder)

    # CSV sims: named "EONR_<sim>.csv"
    csv_files = glob.glob(os.path.join(folder_path, 'EONR_*.csv'))
    for f in csv_files:
        df = pd.read_csv(f)
        rmse = np.sqrt(np.mean((df['pred'] - df['true'])**2))
        sim_name = os.path.splitext(os.path.basename(f))[0]
        records.append({
            'Model': title,
            'Sim': sim_name,
            'RMSE': rmse
        })

    # RDS sims (CF folders): named "sim_<n>.rds"
    rds_files = glob.glob(os.path.join(folder_path, 'sim_*.rds'))
    for f in rds_files:
        result = pyreadr.read_r(f)
        df = next(iter(result.values()))
        # rename CF outputs to match
        df = df.rename(columns={'opt_N_hat':'pred','opt_N':'true'})
        rmse = np.sqrt(np.mean((df['pred'] - df['true'])**2))
        sim_name = os.path.splitext(os.path.basename(f))[0]
        records.append({
            'Model': title,
            'Sim': sim_name,
            'RMSE': rmse
        })

# assemble into a DataFrame
sim_rmse_df = pd.DataFrame(records)

# you can now:
#  - view all 500 RMSEs per model
#  - compute the modelâ€‘level average via:
model_summary = (
    sim_rmse_df
    .groupby('Model')['RMSE']
    .agg(['mean','std','count'])
    .rename(columns={'mean':'Mean_RMSE','std':'SD_RMSE','count':'NSims'})
    .reset_index()
)

# after building `model_summary` as before...

# 1. Sort ascending by Mean_RMSE
model_ranking = model_summary.sort_values('Mean_RMSE').reset_index(drop=True)

# 2. Add a Rank column
model_ranking['Rank'] = model_ranking.index + 1

# 3. Reorder columns for clarity
model_ranking = model_ranking[['Rank', 'Model', 'Mean_RMSE', 'SD_RMSE', 'NSims']]

print(model_ranking)


#print(sim_rmse_df)    # per-sim RMSEs
#print(model_summary)  # per-model summary stats
```
